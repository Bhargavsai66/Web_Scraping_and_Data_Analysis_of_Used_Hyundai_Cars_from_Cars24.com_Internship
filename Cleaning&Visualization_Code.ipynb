```python
import pandas as pd
import re

def clean_location(loc):
    if pd.isna(loc):
        return None

    # Lowercase for uniformity
    loc = str(loc).lower()

    # Remove extra spaces
    loc = re.sub(r'\s+', ' ', loc).strip()

    # If comma exists, take last part (usually city)
    if ',' in loc:
        loc = loc.split(',')[-1].strip()

    # Otherwise, try to take last 1-2 words (city/area)
    else:
        words = loc.split()
        if len(words) > 1:
            loc = ' '.join(words[-2:])  # last two words
        else:
            loc = words[0]

    # Capitalize each word
    loc = loc.title()

    return loc

df['Clean_Location'] = df['Location'].apply(clean_location)
```

```python
df.drop(columns='Location',inplace=True)
df['Location']=df['Clean_Location']
df.drop(columns='Clean_Location',inplace=True)
```

```python
# Remove any non-numeric characters and convert to float
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
# Remove commas and non-digit characters, then convert to numeric
df['KM driven'] = df['KM driven'].astype(str).str.replace(',', '').str.extract('(\d+)')[0]
df['KM driven'] = pd.to_numeric(df['KM driven'], errors='coerce')
```

```python
sns.violinplot(
    data=df,
    x='KM driven',
    y='Transmission',
    inner='stick',
    palette='Dark2'
)

sns.despine(top=True, right=True, bottom=False, left=True)


plt.title('Kilometers Driven Distribution by Transmission Type')
plt.tight_layout()
plt.show()
```

```python
avg_km = df.groupby('Location')['KM driven'].mean().sort_values()


plt.figure(figsize=(10, 6))
sns.barplot(x=avg_km.values, y=avg_km.index, palette='Blues_r')
plt.title('Average Kilometers Driven by Location')
plt.xlabel('Average KM Driven')
plt.ylabel('Location')
plt.tight_layout()
plt.show()
```

```python
location_counts = df['Location'].value_counts().reset_index()
location_counts.columns = ['Location', 'Count']
heatmap_data = location_counts.pivot_table(index='Location', values='Count')

plt.figure(figsize=(6, 8))
sns.heatmap(heatmap_data, annot=True, cmap='YlOrBr', fmt='d')
plt.title('Number of Cars per Location')
plt.tight_layout()
plt.show()
```

```python
avg_km_transmission = df.groupby('Transmission')['KM driven'].mean().sort_values()


plt.figure(figsize=(8, 4))
sns.barplot(x=avg_km_transmission.values, y=avg_km_transmission.index, palette='Set2')
plt.title('Average Kilometers Driven by Transmission Type')
plt.xlabel('Average KM Driven')
plt.ylabel('Transmission')
plt.tight_layout()
plt.show()
```

```python
df['Transmission_num'] = df['Transmission'].astype('category').cat.codes
jitter = np.random.normal(0, 0.1, size=len(df))


plt.figure(figsize=(10, 6))
plt.scatter(df['Transmission_num'] + jitter, df['KM driven'], alpha=0.6, c='teal', edgecolors='w')

plt.xticks(ticks=range(len(df['Transmission'].unique())), labels=df['Transmission'].unique())
plt.xlabel('Transmission Type')
plt.ylabel('Kilometers Driven')
plt.title('Scatter Plot: KM Driven vs Transmission')
plt.tight_layout()
plt.show()
```
